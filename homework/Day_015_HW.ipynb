```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# 原本隨機0.0 ~ 1.0之間,減0.5乘兩倍 -0.5 ~ 0.5 -> -1.0 ~ 1.0
matrix = (np.random.rand(10, 10) - 0.5) * 2
print(matrix)

plt.figure(figsize=(10,10))
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1.0, annot = True, vmax = 1.0)
plt.show()
```

```python
nrow = 1000
ncol = 3

matrix = (np.random.rand(nrow, ncol) - 0.5) * 2

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3

# 隨機常態分布
# randn是normal的一個特殊情況 μ=0, σ=1。均值為0，方差為1。
# np.random.normal(size=(nrow * ncol)),normal可以指定這兩個參數。高斯分佈，拉普拉斯分佈，t分佈都是用來描述樣本特性的
matrix = np.random.randn(nrow,ncol)# 轉成nrow * ncol的陣列?

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
